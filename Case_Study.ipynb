```python
plt.figure(figsize=(9,3))
plt.scatter(df['CGPA'],df['PlacementStatus'])
plt.xlabel("CGPA")
plt.ylabel("PlacementStatus")
plt.title("CGPA vs PlacementStatus")
plt.show()
```

```python
def bar_plot(variable):
    var=df[variable]
    varValue=var.value_counts()
    plt.figure(figsize=(5,3))
    plt.bar(varValue.index,varValue,color="skyblue",edgecolor="black")
    plt.xticks(varValue.index,varValue.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
    print("{}: \n {}".format(variable,varValue))
```

```python
columns_plot=['Internships', 'Projects','Workshops/Certifications','ExtracurricularActivities', 'PlacementTraining', 'PlacementStatus']
for i in columns_plot:
    bar_plot(i)
```

```python
def hist_plot(variable):
  variables=df[variable].value_counts()
  plt.figure(figsize=(15,5))
  plt.hist(variables.index,weights=variables.values,color="skyblue",edgecolor="black")
  plt.xticks(variables.index,variables.index.values)
  plt.xlabel(variable)
  plt.ylabel('Frequency')
  plt.title(variable)
  plt.show()
  print("{}: \n {}".format(variables,variables))
```

```python
columns_plot=['CGPA', 'AptitudeTestScore', 'SoftSkillsRating',  'SSC_Marks','HSC_Marks',]
for i in columns_plot:
    hist_plot(i)
    plt.show()
```

```python
df["PlacementStatus"] = df["PlacementStatus"].str.strip().str.lower()
df["PlacementStatus"] = df["PlacementStatus"].map({"placed": 1, "notplaced": 0})
```

```python
df["ExtracurricularActivities"] = df["ExtracurricularActivities"].map({"Yes": 1, "No": 0})
df["PlacementTraining"] = df["PlacementTraining"].map({"Yes": 1, "No": 0})
```

```python
df.drop(columns=["StudentID"], inplace=True)
```

```python
sns.boxplot(x="PlacementStatus", y="CGPA", data=df)
plt.show()
```

```python
columns_plot=["Internships","Projects","Workshops/Certifications"]
for i in columns_plot:
    sns.countplot(x=i, hue="PlacementStatus", data=df)
    plt.show()

```
